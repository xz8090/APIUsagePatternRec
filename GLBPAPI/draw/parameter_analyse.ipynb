```python
import matplotlib.pyplot as plt
from pylab import *                                 #支持中文
mpl.rcParams['font.sans-serif'] = ['SimHei']


x = [5,10,15,20,25,30,35,40]
y = [0.245407, 0.249445, 0.247404, 0.243623, 0.246083, 0.249476, 0.245519, 0.246619]

plt.xlim(0, 45)  # 限定横轴的范围
plt.ylim(0.24, 0.25)  # 限定纵轴的范围
plt.plot(x, y, marker='o')
plt.legend()  # 让图例生效
plt.xlabel("minsup") #X轴标签
plt.ylabel("MAP") #Y轴标签
plt.title("") #标题
plt.grid()  # 生成网格
plt.savefig('minsup.png', dpi=100)
plt.show()
```

```python
#采样大小和训练时间
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np

mpl.rcParams["font.sans-serif"] = ["SimHei"]
mpl.rcParams["axes.unicode_minus"] = False

x1 = [10,20,30,40]
x2 = [12,22,32,42]
y1 = [549, 970, 1423, 1805]
y2 = [832, 970, 1310, 1530]



plt.bar(x1, y1, 2, align="center", label="1st sample", alpha=0.5)
plt.bar(x2, y2, 2, align="center", label="2st sample", alpha=0.5)

plt.xlabel("size")
plt.ylabel("train time(s)")

plt.xlim(8, 47)  # 限定横轴的范围
plt.ylim(500, 1850)  # 限定纵轴的范围

plt.legend()
plt.grid()  # 生成网格
plt.savefig('sample-time.png', dpi=100)
plt.show()
```

```python
#隐藏层大小和MAP的关系
x = [32,64,96,128]
y1 = [0.268375, 0.271374, 0.263250, 0.280183]
y2 = [0.264228, 0.280183, 0.267314, 0.277958]
#plt.xlim(30, 514)  # 限定横轴的范围
plt.ylim(0.26, 0.285)  # 限定纵轴的范围
tick_label = ["32", "64", "128", "256"]
plt.plot(x, y1, marker='o', label="1st layer")
plt.plot(x, y2, marker='o', label="2st layer")
plt.legend()  # 让图例生效
plt.xlabel("size") #X轴标签
plt.ylabel("MAP") #Y轴标签
plt.title("") #标题
plt.xticks(x, tick_label)
plt.legend()
plt.grid()  # 生成网格

plt.savefig('layer-MAP.png', dpi=100)
plt.show()
```

```python
#隐藏层大小和训练时间
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np

mpl.rcParams["font.sans-serif"] = ["SimHei"]
mpl.rcParams["axes.unicode_minus"] = False

x1 = [10,20,30,40]
x2 = [12,22,32,42]
y1 = [1405, 1407, 1408, 1425]
y2 = [1553, 1425, 1598, 1587]



plt.bar(x1, y1, 2, align="center", label="1st layer", alpha=0.5)
plt.bar(x2, y2, 2, align="center", label="2st layer", alpha=0.5)

plt.xlabel("size")
plt.ylabel("train time(s)")

#plt.xlim(30, 60)  # 限定横轴的范围
plt.ylim(500, 1850)  # 限定纵轴的范围
tick_label = ["32", "64", "128", "256"]
plt.xticks(x1, tick_label)
plt.legend()
plt.grid()  # 生成网格
plt.savefig('layer-time.png', dpi=100)
plt.show()
```
